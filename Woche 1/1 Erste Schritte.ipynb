# Was ist eigentlich ein Jupyter Notebook?

Das Ding, das du hier gerade siehst, ist keine schnöde, statische Webseite. Nee, das ist eine **interaktive Umgebung**, in der du direkt Code schreiben und ausführen kannst. Ziemlich cool, oder?

Jupyter Notebooks – wir nennen sie einfach Notebooks – bestehen aus einzelnen Bausteinen, den **Zellen**. Diese Zellen können im Grunde zwei Dinge sein:

1.  **Text-Zellen (Markdown):** So wie diese hier. Hier kannst du formatierten Text, Bilder, Links, Formeln (LaTeX!) und so weiter einfügen, um deinen Code zu erklären und zu dokumentieren.
2.  **Code-Zellen:** Hier kommt der spannende Teil rein – dein Python-Code (oder auch Code in anderen Sprachen wie R oder Julia, aber wir bleiben hier bei Python).

## Code-Zelle – Hier passiert die Magie!

Unten siehst du ein Beispiel für eine **Code-Zelle**. Da steht ein kurzes Python-Skript drin, das eine simple Berechnung macht, das Ergebnis in einer Variable ablegt und dann ausgibt. Probier's gleich mal aus!


```python
seconds_in_a_day = 24 * 60 * 60
seconds_in_a_day
```

Um den Code in der Zelle oben auszuführen, klick erstmal auf die Zelle, um sie auszuwählen (sie bekommt dann einen Rahmen).

*   **Google Colab:** Drück einfach auf den `▶️` Play-Button links neben dem Code.
*   **Lokale Installation / myBinder:** Nachdem du die Zelle ausgewählt hast, klick oben in der Menüleiste auf `Cell` und dann `Run Cells` oder such nach einem `▶️ Run` Button in der Toolbar. Geht meist schneller!

**Pro-Tipp für alle Umgebungen:** Die Tastenkombination `Strg+Enter` (Windows/Linux) bzw. `Cmd+Enter` (Mac) führt die ausgewählte Zelle ebenfalls aus. `Shift+Enter` führt die Zelle aus UND springt zur nächsten Zelle (oder erstellt eine neue, wenn du am Ende bist) – super praktisch, um schnell durchs Notebook zu gehen.

Willst du den Code ändern? Kein Problem! Klick einfach in die Zelle und tipp los.

### Was bedeuten `In [ ]:` und `In [1]:`?

Code-Zellen haben links so eine Art "Eingabeaufforderung", die dir ein paar Infos gibt:

*   `In [ ]:` (leer): Diese Zelle wurde in der aktuellen Sitzung (seit du das Notebook geöffnet oder den Kernel das letzte Mal neu gestartet hast) noch **nicht ausgeführt**.
*   `In [*]:` (Sternchen): Diese Zelle wird **gerade ausgeführt**. Wenn das Sternchen sehr lange bleibt, könnte dein Code in einer Endlosschleife festhängen oder der Kernel hat sich verabschiedet. Dann musst du vielleicht den Kernel unterbrechen oder neu starten (mehr dazu in der FAQ-Datei).
*   `In [1]:` (Zahl): Diese Zelle wurde ausgeführt. Die Zahl `1` zeigt an, dass es die erste ausgeführte Zelle in dieser Sitzung war. Die nächste wäre `In [2]:` und so weiter. Das ist nützlich, um die Reihenfolge nachzuvollziehen, in der Code ausgeführt wurde.

**Achtung:** Wenn eine Zelle eine Nummer anzeigt (z.B. `In [5]:`), heißt das, sie *wurde irgendwann mal ausgeführt*. Es kann aber sein, dass du den Kernel zwischenzeitlich neu gestartet hast. Dann sind die Variablen aus dieser Zelle vielleicht nicht mehr im Speicher, obwohl die alte Ausgabe noch da steht. Im Zweifel: Zelle einfach nochmal ausführen!

### Ausgabe vs. Anzeige – Ein kleiner, aber feiner Unterschied

Unter einer Code-Zelle können Dinge erscheinen. Manchmal ist das ein bisschen verwirrend, aber es gibt einen Unterschied:

*   **Anzeige-Bereich:** Alles, was dein Code explizit ausgibt, z.B. mit `print("Hallo Welt")` oder wenn du Matplotlib-Plots erzeugst. Das sind Dinge, die der Code aktiv "produziert", um sie dir zu zeigen.
*   **Ausgabe-Bereich (oder "Rückgabe-Bereich"):** Das ist reserviert für das Ergebnis der **letzten Anweisung** in deiner Zelle, *wenn* diese Anweisung einen Wert zurückgibt und du ihn nicht einer Variable zuweist. Im Beispiel oben `seconds_in_a_day` steht die Variable einfach in der letzten Zeile. Jupyter ist so nett und zeigt dir dann den Wert dieser Variable an. Hättest du geschrieben `xyz = seconds_in_a_day`, gäbe es keine direkte Ausgabe, weil das Ergebnis der Zuweisung zugewiesen wird.

Deshalb siehst du oft in Notebooks, dass am Ende einer Zelle einfach nur ein Variablenname steht – das ist der Trick, um sich den Wert schnell anzeigen zu lassen, ohne `print()` benutzen zu müssen.

Variablen, die du in einer Zelle definierst (z.B. `seconds_in_a_day`), kannst du ganz normal in späteren Zellen weiterverwenden, solange der Kernel läuft und nicht neu gestartet wurde:

```python
seconds_in_a_week = 7 * seconds_in_a_day
seconds_in_a_week
```

Du siehst also: Notebooks sind echt mächtig! Du kannst hier **ausführbaren Code**, schick formatierten **Rich Text** (wie diesen hier), **Bilder**, **HTML-Schnipsel**, mathematische Formeln mit **LaTeX**, und noch viel mehr in einem einzigen Dokument kombinieren. Das macht sie zum idealen Werkzeug für Data Science, um Analysen nachvollziehbar zu dokumentieren und Ergebnisse ansprechend zu präsentieren.

## Data Science – Zeit für coole Bibliotheken!

Jetzt wird's richtig interessant! Mit Jupyter Notebooks kannst du die volle Power der genialen Python-Bibliotheken für Data Science nutzen, um Daten zu analysieren und zu visualisieren. Die Code-Zelle unten nutzt zum Beispiel:

*   **NumPy:** Eine super wichtige Bibliothek für alles, was mit Zahlen, Arrays und Matrizen zu tun hat (quasi das Fundament für vieles im Data Science Bereich).
*   **Matplotlib:** Die Standard-Bibliothek in Python, um Diagramme und Visualisierungen zu erstellen.

Klick einfach auf die Zelle, um den Code anzuschauen oder zu ändern. Und dann: Ausführen und staunen!

```python
%pip install numpy matplotlib
```

```python
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Sample Visualization")
plt.show()
```
